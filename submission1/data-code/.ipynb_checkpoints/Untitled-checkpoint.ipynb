```r
# Set CRAN repository
options(repos = c(CRAN = "https://cloud.r-project.org"))

# Install and load required packages
if (!require("pacman")) install.packages("pacman")
pacman::p_load(tidyverse, ggplot2, dplyr, lubridate, stringr, readxl, data.table, gdata, scales)
source("functions.R")
# Define variables
monthlist <- sprintf("%02d", 1:2)  # Generates a list of months as "01", "02"
y <- 2014  # Year for analysis
```

```r
# Function to load enrollment data for a specific month and year
load_month <- function(month, year) {
  file_path <- paste0("/home/salizad/econ470/a0/work/ma-data/ma/enrollment/Extracted Data/",
                       "CPSC_Enrollment_Info_", year, "_", month, ".csv")
     # Check if the file exists
  if (!file.exists(file_path)) {
    stop(paste("File does not exist:", file_path))
  }
  
  return(read_csv(file_path, show_col_types = FALSE))
}
```

```r
months <- sprintf("%02d", 1:12)
years <- 2014:2019
```

```r
yearly_summaries <- list()

# Loop through each year 
for (current_year in years) {
    enrollment_data <- map_dfr(months, ~ load_month(.x, current_year))
    enrollment_data <- enrollment_data %>%
        filter(!grepl("SNP|800|PDP", `Plan ID`))
}
```

```r
# Check for missing values
summary(enrollment_data)
```

```r
# Summarize data for the current year 
plan_summary <- enrollment_data %>%
        group_by(`County`) %>%
        summarise(plan_count = n(), .groups = "drop") %>%
        mutate(year = current_year) %>%
        ungroup()
```

```r
# Store the yearly summary 
yearly_summaries[[as.character(current_year)]] <- plan_summary

# Combine all yearly summaries into a single data frame
final_summary <- bind_rows(yearly_summaries)
```

```r
# Specify the output directory
output_dir <- "/home/salizad/econ470/a0/work/Homework2/Data/output"

# Create the directory structure if it doesn't exist
dir.create(output_dir, recursive = TRUE, showWarnings = FALSE)

# Output results to the console
if (nrow(final_summary) == 0) {
    cat("No data loaded.\n")
} else {
    cat("Data loaded successfully:\n")
    print(head(final_summary))  # Show the first few rows
    
    # Save to the specified CSV file
    write.csv(final_summary, file.path(output_dir, "enrollment_data_summary.csv"), row.names = FALSE)
    cat("Summary saved to 'Homework2/Data/output/enrollment_data_summary.csv'.\n")
}
```

```r
# Boxplot of plan counts by year
ggplot(final_summary, aes(x = factor(year), y = plan_count)) +
    geom_boxplot() +
    labs(title = "Distribution of Plan Counts by County (2014-2019)",
         x = "Year", y = "Number of Plans")

# Specify the output directory for visualizations
output_analysis_dir <- "/home/salizad/econ470/a0/work/Homework2/submission1/analysis"

# Create the directory structure if it doesn't exist
dir.create(output_analysis_dir, recursive = TRUE, showWarnings = FALSE)

# Output results to the console
if (nrow(final_summary) == 0) {
    cat("No data loaded.\n")
} else {
    cat("Data loaded successfully:\n")
    print(head(final_summary))  # Show the first few rows
    
    # Save to the specified CSV file
    write.csv(final_summary, file.path(output_dir, "enrollment_data_summary.csv"), row.names = FALSE)
    cat("Summary saved to 'Homework2/Data/output/enrollment_data_summary.csv'.\n")
    
    # Visualization: Boxplot of plan counts by year
    enrollment_plot <- ggplot(final_summary, aes(x = factor(year), y = plan_count)) +
        geom_boxplot() +
        labs(title = "Distribution of Plan Counts by County (2014-2019)",
             x = "Year", y = "Number of Plans")
    
    # Save the plot to the specified analysis directory
    ggsave(filename = file.path(output_analysis_dir, "enrollment_boxplot.png"), plot = enrollment_plot)
    cat("Boxplot saved to 'Homework2/submission1/analysis/enrollment_boxplot.png'.\n")
}
```
